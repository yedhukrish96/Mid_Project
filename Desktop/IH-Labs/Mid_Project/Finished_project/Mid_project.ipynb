# Dealing with NaNs in Overall Rating column so as to not drop the rows as this column in crucial for me.

```python



def calculate_overall_rating(customer_satisfaction_rate):
    if 58 <= customer_satisfaction_rate <= 60:
        return 100
    elif 54 <= customer_satisfaction_rate <= 57:
        return 90
    elif 50 <= customer_satisfaction_rate <= 53:
        return 88
    elif 46 <= customer_satisfaction_rate <= 49:
        return 84
    elif 40 <= customer_satisfaction_rate <= 45:
        return 80
    elif 34 <= customer_satisfaction_rate <= 39:
        return 75
    elif 28 <= customer_satisfaction_rate <= 33:
        return 70
    elif 20 <= customer_satisfaction_rate <= 27:
        return 65
    elif 10 <= customer_satisfaction_rate <= 19:
        return 60
    else:
        return np.nan

# Apply the function to fill NaN values in 'Overall Ratings' column
data['Overall Rating'] = data['Overall Rating'].fillna(data['Customer Satisfaction Rate_out_of_60'].apply(calculate_overall_rating))
```

# Dealing with NaNs in my Postal Code column

```python
# s= str(10115.0)
# s.split('.')[0]

def fill_postal_code(postal_code, neighbourhood_group):
    group_postal_code = {
        'Mitte': '10178',
        'Friedrichshain-Kreuzberg': '10247',
        'Pankow': '13156',
        'Neukölln': '12047',
        'Charlottenburg-Wilm.': '10585',
        'Tempelhof - Schöneberg': '10823',
        'Lichtenberg': '10317',
        'Treptow - Köpenick': '12559',
        'Steglitz - Zehlendorf': '14169',
        'Reinickendorf': '13403',
        'Marzahn - Hellersdorf': '12679',
        'Spandau': '13581'
    }
    
   
    if pd.isna(postal_code):
        if neighbourhood_group == 'Mitte':
            postal_code = '10178'
        elif neighbourhood_group == 'Friedrichshain-Kreuzberg':
            postal_code = '10247'
        elif neighbourhood_group == 'Pankow':
            postal_code = '13156'
        elif neighbourhood_group == 'Neukölln':
            postal_code = '12047'
        elif neighbourhood_group == 'Charlottenburg-Wilm.':
            postal_code = '10585'
        elif neighbourhood_group == 'Tempelhof - Schöneberg':
            postal_code = '10823'
        elif neighbourhood_group == 'Lichtenberg':
            postal_code = '10317'
        elif neighbourhood_group == 'Treptow - Köpenick':
            postal_code = '12559'
        elif neighbourhood_group == 'Steglitz - Zehlendorf':
            postal_code = '14169'
        elif neighbourhood_group == 'Reinickendorf':
            postal_code = '13403'
        elif neighbourhood_group == 'Marzahn - Hellersdorf':
            postal_code = '12679'
        else:
            postal_code = '13581'
    
    return postal_code

data['Postal Code'] = data.apply(lambda row: fill_postal_code(row['Postal Code'], row['Neighborhood Group']), axis=1)

```

```python
def remove_decimal(postal_code):
    if isinstance(postal_code, float):
        postal_code = str(postal_code)
    return postal_code.split(".")[0]


data["Postal Code"] = data["Postal Code"].apply(remove_decimal)
```

```python
def remove_newline(postal_code):
     return postal_code.split("\n")[0]
    
data['Postal Code']= data['Postal Code'].apply(remove_newline)
```

# Final cleaning up of my dataset

```python
def boolean_superhost(is_superhost):
    if is_superhost == 'f':
        return "False"
    else:
        return "True"
    
data['Is Superhost'] = data['Is Superhost'].apply(boolean_superhost)
```

```python
def boolean_location(is_exact_location):
    if is_exact_location == 'f':
        return "False"
    else:
        return "True"
    
data['Is Exact Location'] = data['Is Exact Location'].apply(boolean_location)
```

```python
def boolean_business(business_travel_ready):
    if business_travel_ready == 'f':
        return "False"
    else:
        return "True"
    
data['Business Travel Ready'] = data['Business Travel Ready'].apply(boolean_location)
```

```python
def boolean_instant(instant_bookable):
    if instant_bookable == 'f':
        return "False"
    else:
        return "True"
    
data['Instant Bookable'] = data['Instant Bookable'].apply(boolean_location)
```

```python
from scipy.stats import pearsonr
import scipy.stats as stats
```

```python
x = final_df['Price In_Eur']
y = final_df['Overall Rating']


corr, pval = stats.pearsonr(x, y)

# Print the results
print('Correlation coefficient:', corr)
print('P-value:', pval)

# Define the null and alternative hypotheses
null_hypothesis = "There is no correlation between price and overall rating."
alternative_hypothesis = "There is a positive correlation between price and overall rating."

# Set the significance level
alpha = 0.05

# Determine whether to reject or fail to reject the null hypothesis
if pval < alpha:
    print("Reject the null hypothesis.")
    print(alternative_hypothesis)
else:
    print("Fail to reject the null hypothesis.")
    print(null_hypothesis)
```
